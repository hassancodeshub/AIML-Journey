```python
import requests
from bs4 import BeautifulSoup
```

```python
page_count = 1

while True:

    url = f"https://quotes.toscrape.com/page/{page_count}/"
    
    res = requests.get(url)

    soup = BeautifulSoup(res.text,"lxml") # To find the end of the webpage

    quote  = soup.find_all("div", class_="quote")

    if not quote:
        print("No Pages ahead!")
        break

    with open(f"page{page_count}.html","w",encoding="utf-8") as f:
        f.write(res.text)

    print(f"Page {page_count} Exracted Successfully!")

    page_count+=1
```

```python
op=[]

for i in range(1,11):
    with open(f"page{i}.html","r",encoding="utf-8") as f:
        data = f.read()
        soup = BeautifulSoup(data,"lxml")

    all_quotes = soup.select("div.quote")  # soup.findall("div",class_="quote")

    for q in all_quotes:
        tags = [tag.strip() for tag in q.find("meta",class_="keywords")["content"].split(",")]

        if "life" in tags:
            l=[]
            l.append(q.find("span",class_="text").get_text())
            l.append(q.find("small",class_="author").get_text())
            op.append(l)
            print(l)
            

    print(f"=========={i}=========")

print(len(op))
```

```python
import pandas as pd

df = pd.DataFrame(op, columns=["Quotes","Authors"])
df
```
