# B :

```python
# B. Select only those rows where:
# • petal_length > 4.5

print(d.query("petal_length>4.5"))

print(d[d["petal_length"] > 4.5])
```

```python
# • species = "Iris-virginica"
var = "Iris-virginica"
print(d.query("species==@var"))

print(d[d["species"] == "Iris-virginica"])
```

# C. Group by species and compute:

```python
# • Average sepal_length
d.groupby("species")["sepal_length"].mean()
```

```python
# • Maximum petal_width

d.groupby("species")["petal_width"].max()
```

```python
# • Standard deviation of sepal_width

d.groupby("species")["sepal_width"].std()
```

# D. Create a new column: "petal_ratio" = petal_length / petal_width
• For each species, find the average petal_ratio.


```python
d = d.assign(petal_ratio = d["petal_length"]/d["petal_width"])
print(d)

print(f"For each sepcies average petal ratio :")

print(d.groupby("species")["petal_ratio"].mean())
```
