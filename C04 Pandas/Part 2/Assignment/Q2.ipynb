# C. Group by Pclass and compute:
• Survival rate
• Average Fare
• Average Age


```python
result  = d.groupby("Pclass").agg(
        survival_rate = ("Survived","mean"),
        average_fare = ("Fare","mean"),
        average_age = ("Age","mean")
)

print(result)
result["survival_rate"] = result["survival_rate"] * 100
print(result)
```
