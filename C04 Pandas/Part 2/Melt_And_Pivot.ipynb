```python
import pandas as pd
d = pd.read_csv("raw_data.csv")
```

```python
d
```

```python
# 1] Melting : [Wide to long]

d = d.melt(
    id_vars = "name",
    value_vars = ["country","gender"],
    var_name = "Melted_Col_Heading",
    value_name = "Melted_Values"
)
```

```python
d = d.drop_duplicates()
print(d)
```

```python
# 2] Pivot : [Long to Wide]

d = d.pivot(
    index = "name",
    columns = "Melted_Col_Heading",
    values = "Melted_Values"
)
```
