```python
import pandas as pd
```

```python
d = pd.read_csv("raw_data.csv")
```

```python
# 1. apply(function)

# A new column will be added
d["tax"] = d["income"].apply(lambda x : "20%" if x>=50000 else "10%")

print(d)
def func(x):
    return ("True" if int(x)>=50000 else "False")
d["income"]=d["income"].ffill()
d["income"] = d["income"].apply(func)
print(d)
```

```python
# map function :
dic={
    "True" : "T",
    "False" : "F"
}
d["income"] = d["income"].map(dic)

print(d)
```

```python
# 3] Assign Function : Creates a new column.

d = d.assign(new_age  = d["age"]*2)
print(d)
```

```python
# 4] d.repalce(old,new)

d["name"] =  d["name"].replace("Alex","alooo")
print(d)
```

```python
d.columns = ["A","B","C","D","E","F","G","H"]
print(d)
```

```python
d.rename(columns={"Azzzz" : "zzz"})
```

```python
d.rename(index={1 : "Faaaah"})
```

```python
print(d["C"].sort_values())
print(d.sort_values("C",ascending = False))
print(d.sort_values(["C","H"]))
```

```python
dc=d.sort_values(["C","H"])
print(dc)

print(dc.sort_index())
```

```python
# Reset Index :

dc = dc.sort_values("H")
print(dc)

# dc = dc.reset_index() # Will print a old index column
# print(dc)

dc.reset_index(drop = True,inplace = True)
print(dc)
```

```python
n = pd.read_csv("raw_data.csv")
print(n)

n["rank"] = n["income"].rank()
print(n)



n["rank"] = n["income"].rank(ascending = False,method = "dense")
print(n)

n["rank"] = n["income"].rank(ascending = False,method = "min")
print(n)

n["rank"] = n["income"].rank(ascending = False,method = "max")
print(n)
```
