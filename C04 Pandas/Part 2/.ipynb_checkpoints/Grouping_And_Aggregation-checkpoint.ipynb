```python
import pandas as pd
d = pd.read_csv("raw_data.csv")
```

```python
d.groupby("country")["income"].mean()
```

```python
# Applying Multiple Functions :

print(d.groupby("country")["income"].agg(["mean","max","sum"]))
print(d.groupby("country")["income"].aggregate(["mean","max","sum"]))


# With custom labels :
print(d.groupby("country")["income"].agg(avg_mean="mean",min_val="min"))
```

```python
# Applying diff functions to diff columns ;

d.groupby("country").agg(
    {
        "income":"mean",
        "age":"max"
    }
)
```

```python
# Applying diff functions to diff columns ; (Custom Labels))

d.groupby("country").agg(
    avg_sal = ("income",'mean'),
    new = ("age","max")
)
```
