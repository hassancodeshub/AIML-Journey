```python
import pandas as pd
```

```python
d1 = pd.DataFrame(
    {
        'name' : ["Atiqa","Sadiqa","Saniya"],
        'id' : [1,3,3],
        'val' : ['a','b','c']
    }
)

d2 = pd.DataFrame(
    {
        'name' : ["Ali","Hun","Rose"],
        'id' : [1,2,2],
        'val' : ['f','e','d']
    }
)

print(d1)

print(d2)
```

```python
# 1] Inner Join :

pd.merge(d1,d2,on="id")
```

```python
# 2] Outer Join :

pd.merge(d1,d2,on="id",how="outer")
```

```python
# 3] Left Join ;

pd.merge(d1,d2,on="id",how="left")
```

```python
# 3] Right Join ;

pd.merge(d1,d2,on="id",how="right")
```
