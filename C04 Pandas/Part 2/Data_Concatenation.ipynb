```python
import pandas as pd
```

```python
d1 = pd.DataFrame(
    {
        'name' : ["Atiqa","Sadiqa","Saniya"],
        'id' : [1,3,3],
        'val' : ['a','b','c']
    }
)

d2 = pd.DataFrame(
    {
        'name' : ["Ali","Hun","Rose"],
        'id' : [1,2,2],
        'val' : ['f','e','d']
    }
)

print(d1)

print(d2)
```

```python
# 1] Row-Wise :

pd.concat([d1,d2])
```

```python
# 2] Col-Wise :

pd.concat([d1,d2],axis=1)
```

```python
# 3] Ignoring the previous idex :

pd.concat([d1,d2],ignore_index=True)
```
