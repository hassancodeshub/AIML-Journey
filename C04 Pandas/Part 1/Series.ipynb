```python
import pandas as pd
```

```python
s = pd.Series([10,20,30,40,50])
print(f"This is the Series : [index :  value ]")
print(s)

# Can be accessed by label (index):
print(s[0])
print(s.index)
```

```python
s = pd.Series([9.43,8.0,7.9],index=["Hassan","Sumit","Gaurav"])
print(f"This is the Series : [Custon label :  value ]")
print(s)

# Can be accessed by Label as well as index :
print(s["Hassan"])
# print(s[0]) It will throw a warning

print(f"All indices : {s.index}")
```
