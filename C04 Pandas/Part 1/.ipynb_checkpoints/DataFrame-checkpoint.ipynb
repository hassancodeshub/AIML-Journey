```python
import pandas as pd
import numpy as np
```

```python
# Creation Using The Dicitonary :
d  = {
    "Head_1" : [1,2,3],
    "Head_2" : [4,5,6],
    "Head_3" : [7,8,9]
}

f1 = pd.DataFrame(d , index=["H1","H2","H3"])
print("Each Key As FIeld and Each Value as Column of Key :")
f1
print(f1.index)
print(f1.columns)
```

```python
# Creation Using The List Of Lists ;
f2 = pd.DataFrame([["Row 1",2],["Row 2",4]])
print(f2)
f3 = pd.DataFrame([["Row 1",2],["Row 2",4]],columns = ["C1","C2"])
print(f3)
f4 = pd.DataFrame([["Row 1",2],["Row 2",4]],columns = ["C1","C2"],index = ["I1","I2"])
print(f4)
```

```python
# Using the NumPy Arrays L

arr = np.array([[10,20,30],[40,50,60]])

d  = pd.DataFrame(arr)
print(d)
print(d.index)
print(d.columns)

d2 = pd.DataFrame(arr,index = ["A","B"],columns = ["H1","H2","H3"]) 
print(d2)
```
