```python
import pandas as pd
d = pd.read_csv("raw_data.csv")
```

```python
d
```

```python
# Convert into Lowercase :

print(d["name"].str.lower())

# Convert into Uppercase :

print(d["name"].str.upper())

# First letter capital :

print(d["country"].str.capitalize())

# Remove back and trailing spaces :

print(d["name"].str.strip())


# Check if a value exist in coloumn :

print(d["country"].str.contains("india",case=False))

# Split on the basis of a seprator :

x = d["name"].str.split(" ")
print(list(x))
```
