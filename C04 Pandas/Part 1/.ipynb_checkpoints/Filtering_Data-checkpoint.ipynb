```python
import pandas as pd
d = pd.read_csv("globalAirQuality.csv")
```

```python
# Single Condition :

print(d[d["aqi"] > 100])

# Multiple Condition :

print(d[(d["aqi"] > 100) & (d["pm25"] < 50)])

# Chaining :

print(d[d["aqi"] > 150][["city","pm10","aqi"]])
```

```python
# Using Query : (Written a copy )

print(d.query("aqi>10"))

# Chaining :
print(d.query("aqi > 100")[["aqi"]])

# Multiple condition :
print(d.query("aqi>100 & pm10 >10"))
```
