```python
import pandas as pd
d = pd.read_csv("globalAirQuality.csv")
```

```python
# Column - Wise  :
d["aqi"]
d[["city","aqi"]]
```

```python
# Row - Wise :

# 1] Label Based : d.loc[start : end(includsive)]

print(d.loc[0])

print(d.loc[0 : 2])


# 2] Column Based : data.iloc[start : end(exclusive)]

print(d.iloc[0])

print(d.iloc[0 : 2])
```

```python
# Selecting Row - Col :
d.head(1)

# 1] Using Loc :

print(d.loc[0,"aqi"])

print(d.loc[0 : 1,["no2","co","humidity"]])

# 2] Using Iloc :

print(d.iloc[0,11])

print(d.iloc[0 : 2, [7,11,13]])
```

```python
# Single Cell :


# 1] Using at : (Label Wise )

print(d.at[0,"aqi"])

# 2] Using iat : (Index Wise)

print(d.iat[0,11])
```
