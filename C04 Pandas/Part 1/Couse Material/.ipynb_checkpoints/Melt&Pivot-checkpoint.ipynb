```python
import pandas as pd
```

```python
df  = pd.read_csv("sorted_data.csv")
```

```python
df
```

```python
melted  = pd.melt(df,id_vars=["id","name"],value_vars=["income","gender"],var_name="Meleted",value_name="Values")
print(melted)
```

```python
reverted = pd.pivot(melted,index=["id","name"],columns="Meleted",values="Values")
print(reverted)
print(df)
reverted.reset_index()
```
