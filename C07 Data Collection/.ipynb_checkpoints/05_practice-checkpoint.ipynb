```python
from bs4 import BeautifulSoup

f = open("scraped_data/countries.html")
html_content = f.read()

soup = BeautifulSoup(html_content,"lxml")
```

```python
# Extracting the country names :

tags = soup.find_all('h3')

names = []

for country in tags:
    names.append(str(country.text.strip()))

print(names)
print(len(names))
```

```python
# Extracting the Populations :

pop_tag = soup.find_all('span', class_="country-population")

populations=[]

for p in pop_tag:
    populations.append(int(p.text.strip()))

print(populations)
```

```python
dic ={}

for i,c in enumerate(names):
    dic[c] = populations[i]

print(dic)
```
