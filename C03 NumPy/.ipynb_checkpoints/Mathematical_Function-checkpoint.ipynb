```python
import numpy as np
```

```python

ar = np.array([30,10,20,10,50.64,40])

# Aggregates :

print(ar)
print(np.min(ar))
print(np.max(ar))
print(np.argmin(ar))
print(np.argmax(ar))
print(np.sum(ar))
print(np.prod(ar))
print(np.mean(ar))
print(np.var(ar))
print(np.std(ar))

# Power :
print(np.square(ar))
print(ar ** 2)
print(np.sqrt(ar))
print(np.pow(ar , 3))


# Logarithm :
print(np.log(ar)) # Natural Log
print(np.log10(ar))
print(np.log2(ar))
print(np.exp(ar))

# Round :
print(np.round(ar))
print(np.ceil(ar))
print(np.floor(ar))
print(np.trunc(ar))

#Extra :
print(np.unique(ar))
print(np.abs(ar))
print(np.sort(ar))
```
