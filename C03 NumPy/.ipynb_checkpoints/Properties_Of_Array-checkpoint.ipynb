```python
import numpy as np
```

```python
ar=np.array([1,2,3,4,5])

#(row,col)
print(ar.shape)

#no. of elements
print(ar.size)

#1D,2D...
print(ar.ndim)

#data_type64
print(ar.dtype)
```

```python
#Type Casting:

float_array=ar.astype(np.float64)
print(float_array,float_array.dtype)


new = np.array([1,2,3,4,5],dtype="float64")
print(new,new.dtype)
```
