```python
import numpy as np
```

```python
# Sum of 2 - D Array :

ar = np.array([[1,2],[3,4]])

# Whole Sum :
print(np.sum(ar))

# Col - Sum :
print(np.sum(ar,axis = 0))

# Row - Sum :
print(np.sum(ar,axis = 1))
```

```python
# Slicing :

"""
  0 1 2
0 4 5 6 
1 7 8 9
2 4 2 5
"""

ar=np.array([[4,5,6],[7,8,9],[4,2,5]])
print(ar)

new = ar[0:2,1:2]
print(new)
```

```python
ar2 = np.array([[1,2],[4,3]])
print(ar2)
print(ar2[0][1])
print(ar2[0,1])
```

```python
# 3D Array:

td = np.array([[[0,1],[2,3]],[[4,5],[6,7]]])
print(td,td.ndim,td.shape)

#Access :

print(td[1][1][0])
print(td[1,1,0])

#Slicing using axis ;

print(td[0:2,0:2,0:1])

# Modifying the sliced area with a particular value :

td[0:2,0:2,0:1] = 66
print(td)
```
