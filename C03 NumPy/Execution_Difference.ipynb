```python
import numpy as np
import time as t
import sys
```

```python
#Execution TIme :

size = 10000000
l1=list(range(size))
start = t.time()
sq=[x**2 for x in l1]
end = t.time()
print(end-start)
arr=np.array(l1)
start = t.time()
sq=arr**2
end = t.time()
print(end-start)
```

```python
#Size Difference :

print(f"Size of Python List : {sys.getsizeof(l1)*len(l1)} bytes")
print(f"Size of NumPy Array : {arr.nbytes} bytes")
```
