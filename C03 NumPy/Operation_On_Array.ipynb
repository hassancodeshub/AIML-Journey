```python
import numpy as np
```

```python
ar=np.array([[1,2,3],[4,5,6]])
print(ar,ar.shape)

#Reshaping :
reshaped = ar.reshape((3,2))
print(reshaped,reshaped.shape)

#Converting 2-D Into 1-D :
flat = reshaped.flatten()
print(flat,flat.shape)
```

```python
# Normal Indexing :

print(flat[0])
flat[0]=10
print(flat[0])
```

```python
# Fancy Indexng :

index = [0,2,3]
print(flat[index])
```

```python
# Boolean Indexing :

print(flat[flat % 2 == 0])
print(flat[flat>5])
```

```python
# Array Slicing L

flat=flat[1:4:2]
print(flat)
```

```python
# View vs Copy :

#Python List :

lst = [10,20,30,40,50]
print(f"Orignal List : {lst}")

sub = lst [2 : 5]
print(f"SubList : {sub}")

sub[0] = "x"

print("After :")
print(f"Orignal List : {lst}")
print(f"SubList : {sub} ")



# Numpy Arrays :

arr = np.array(lst)

print(f"Orignal Array : {lst}")

subr = arr [2 : 5]
print(f"SubArray : {subr}")

subr[0] = 100000

print("After :")
print(f"Orignal List : {arr}")
print(f"SubArray : {subr} ")
```
