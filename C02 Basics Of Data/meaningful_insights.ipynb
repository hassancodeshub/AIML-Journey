```python
import json
```

```python
with open ("cleaned_data.json","r") as f:
    data = json.load(f)
```

```python
# 1. Find the average rating :
sum = 0
count = 0
for user in data:
    count+=1
    sum+=float(user["rating"])
avg=sum/count
print(avg)
```

```python
# Percentage of user <3 rating :
u=0
for user in data:
    if(float(user["rating"])<3.0):
        u+=1
percent=(u/count)*100
print(f"{percent} %")
```

```python
# Recommendation :
rec=[]
for user in data:
    dic={}
    if (float(user["rating"])<=4):
        dic[user["name"]]="samsung"
    else :
        dic[user["name"]]="Apple"
    rec.append(dic)
print(rec)
```
