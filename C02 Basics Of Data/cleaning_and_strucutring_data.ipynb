```python
import json
```

```python
#Load The Data :

def load_data(filename):
    with open (filename,"r") as f:
        data=json.load(f)
        return data
        
data=load_data("data.json")
print(data)
```

```python
# Cleaning the Rating Part :

def clean(data):
    num_dic={
        "one" : 1,
        "two" : 2,
        "three" : 3,
        "four" : 4,
        "five" : 5}
    for user in data:
        print(user)
        if user["rating"] in num_dic:
            user["rating"]=str(num_dic[user["rating"]])
        else :
            user["rating"]=(str(user["rating"]).strip())
        print(user)

    return data

cleaned=clean(data)
```

```python
# Filling the Missing data :
def missing(data):
    lst=["name","rating","feedback","age"]
    for user in data:
        for parameter in lst:
            user.setdefault(parameter,None)
    return data

filled=missing(cleaned)
for user in filled:
    print(user)
```

```python
#Removing The Duplicate Values (De-Duplication):

def de_duplication(data):
    l=[]
    u=set()
    for user in data:
        if user["name"].strip().lower() in u:
            continue
        l.append(user)
        u.add(user["name"].strip().lower())
    return l

de_deuplicated = de_duplication(filled)
print(de_deuplicated)
```

```python
with open ("cleaned_data.json","w") as f:
    json.dump(de_deuplicated,f,indent = 4)
```
