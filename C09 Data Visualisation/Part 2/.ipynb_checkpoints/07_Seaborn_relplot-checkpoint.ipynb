```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Default Theme :

sns.set_theme()
```

```python
# Default Data Sets :

sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
print(tips)
print(type(tips))
```

```python
print(tips.sample(10))
print(tips.nunique())
```

```python
sns.relplot(
    data = tips,
    x = "total_bill",
    y = "tip"
)
```

```python
# Custom Theme :

sns.set_theme(style="ticks")

# Styles : Available on websites

sns.relplot(
    data = tips,
    x = "total_bill",
    y = "tip"
)
```

```python
# Highlighting On basis of parameter : (hue)
# Custom Point Size : size=(size_list)
# Custom symbols On basis of parameter : (style=parameter

sns.set_theme()

sns.relplot(
    data = tips,
    x = "total_bill",
    y = "tip",
    hue = "smoker",
    size = "size",
    style="smoker",
)
```

```python
# Line Plot : (Sorts the x values first)

x = [1,2,6,4,5]
y = [1,4,36,16,25]

sns.relplot(
    x=x,
    y=y,
    kind="line"
)
```

```python
# Line Plot : (Maplot Lib Doesn't sort the data )

x = [1,2,6,4,5]
y = [1,4,36,16,25]

plt.plot(
    x,y
)
```

```python
# relplot kind = "scatter or lineplot only"
```

```python
# Subplots using the scatter plots :

sns.relplot(
    data = tips,
    x = "total_bill",
    y = "tip",
    col = "day",
    row = "sex"
)
```
