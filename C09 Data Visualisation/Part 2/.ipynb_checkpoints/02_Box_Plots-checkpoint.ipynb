```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data  = [7,8,5,6,9,4,10,12,15]
print(len(data))
plt.boxplot(data)
plt.grid()
```

```python
# Multiple datasets :

np.random.seed(0)
grp_1 = np.random.normal(50,10,100)
grp_2 = np.random.normal(60,15,100)

plt.boxplot([grp_1,grp_2],tick_labels=["Group 1","Group 2"],showmeans=True)
plt.grid()
plt.show()
```

```python
# Horizontal Box Plot :

data  = [7,8,5,6,9,4,10,12,15]
plt.boxplot(data,vert=False,showmeans=True)  #(showmeans=True) : Display the mean of the dataset
plt.grid()
```

```python
# Custom whisker length :

data  = [7,8,5,6,9,4,10,12,15]
plt.boxplot(data,vert=False,showmeans=True,whis=1.0)
plt.grid()
```
