```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme()
tip = sns.load_dataset("tips")
```

```python
fig,ax = plt.subplots()

sns.lineplot(
    data=tip,
    x="day",
    y="total_bill",
    hue="sex",
    ax=ax,
    marker="o",
    errorbar=None
)

ax.set_title("Average Total Bill by Day and Sex")
ax.set_xlabel("Day of Week")
ax.set_ylabel("Total Bill ($)")

ax.legend(title="Sex")
plt.tight_layout()

plt.show()
```
