```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(0)
scores =  np.random.normal(70,10,100)
plt.hist(scores,color="skyblue",edgecolor="black",bins=10)
plt.title("Normal Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")

# y=0
# for x in scores:
#     if(x>=60 and x<=70):
#        y+=1
# print("Values in between 60 to 70 : ",y)
```

```python
# Custom bins: 

ranges = [0,50,60,70,80,90]

np.random.seed(0)
scores =  np.random.normal(70,10,100)
plt.hist(scores,color="skyblue",edgecolor="black",bins=ranges)
plt.title("Normal Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")

# For custom labels :
# plt.xticks(ranges)
```

```python
# Custom bins: 

marks = [
    25, 38, 42, 55, 61, 67, 72, 78, 83, 88,
    91, 95, 47, 53, 59, 64, 69, 74, 79, 85,
    92, 97, 33, 45, 58, 62, 71, 76, 81, 89
]

bins = [0, 30, 70, 90, 100]

plt.hist(marks,color="skyblue",edgecolor="black",bins=bins)
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
```

```python
# Multiple Data set :

legit_transactions = [
    2.99, 5.49, 8.99, 12.50, 14.99, 19.99, 23.45, 29.99, 34.99, 39.50,
    45.00, 49.99, 55.25, 60.00, 75.99, 89.99, 120.50, 150.00, 199.99,
    249.99, 300.75, 450.00, 600.00, 850.00, 1200.00
]

fraud_transactions = [
    50, 100, 150, 200, 300, 500, 500, 750, 1000, 1000,
    1200, 1500, 1500, 2000, 2500, 3000, 3000
]

# Transaction Amount Distribution - Legit VS Fraud :

print("Legit transactions are left skewed!")

plt.hist(fraud_transactions,color="red",edgecolor="Black",label="Fraud",bins=20,alpha=0.5)
plt.hist(legit_transactions,color="green",edgecolor="Black",label="Legit",bins=20,alpha=0.5)

plt.title("Transaction Amount Distribution - Legit VS Fraud")
plt.xlabel("Amount ($)")
plt.ylabel("Frequency")
plt.legend()
plt.grid()
```

```python
# AXV Lines :  (To highlight a value using a line)

marks = [
    25, 38, 42, 55, 61, 67, 72, 78, 83, 88,
    91, 95, 47, 53, 59, 64, 69, 74, 79, 85,
    92, 97, 33, 45, 58, 62, 71, 76, 81, 89
]

bins = [0, 30, 70, 90, 100]

plt.hist(marks,color="skyblue",edgecolor="black",bins=bins)
plt.axvline(
    x=40,
    label="Passing Marks [40]",
    color="red",
    linestyle="--",
    linewidth=2
)
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.grid()
plt.ylabel("Frequency")
```
