```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme()
flight = sns.load_dataset("flights")
```

```python
flight.head(10)
```

```python
f = flight.pivot(index="month",columns="year",values="passengers")
```

```python
f
```

```python
sns.heatmap(
    f,
    cmap = "coolwarm"
)
```

```python
# To display the value :

sns.heatmap(
    f,
    cmap = "coolwarm",
    annot = True,
    fmt = "d"
)
plt.title("Passengers Data")
plt.tight_layout()
```
