```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
flights = sns.load_dataset("flights")
```

```python
flights.head(12)
```

```python
# Dipsplays the mean/average of the data by default for duplicate :

sns.lineplot(
    data = flights,
    x = "year",
    y = "passengers"
)
```

```python
# To seprate the duplicates using hue :

sns.lineplot(
    data = flights,
    x = "year",
    y = "passengers",
    hue="month"
)
```

```python
# To seprate the duplicates using turning off the estimator:

sns.lineplot(
    data = flights,
    x = "year",
    y = "passengers",
    estimator = None
)
```
