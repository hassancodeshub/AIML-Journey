```python
import matplotlib.pyplot as plt
```

```python
fig , ax = plt.subplots(2,2)

print(type(fig),type(ax))

ax[0][0].plot([1,2,3],[1,2,3])
ax[0][0].set_title("Y = X")
ax[0][0].grid()

ax[0][1].plot(list(map((lambda x : x**2),[1,2,3])),[1,2,3])
ax[0][1].set_title("X ** 2")
ax[0][1].grid()

ax[1][0].plot(list(map((lambda x : x**3),[1,2,3])),[1,2,3])
ax[1][0].set_title("X ** 3")
ax[1][0].grid()

ax[1][1].plot(list(map((lambda x : x**4),[1,2,3])),[1,2,3])
ax[1][1].set_title("X ** 4")
ax[1][1].grid()


fig.tight_layout()
```
