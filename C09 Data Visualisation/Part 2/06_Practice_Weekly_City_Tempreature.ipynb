```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data : Temperature over cities
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]

cities = ["New York", "London", "Delhi", "Tokyo"]

temperatures = [
    [22, 23, 21, 24, 25],   # New York
    [18, 19, 17, 20, 21],   # London
    [30, 32, 31, 33, 34],   # Delhi
    [25, 26, 24, 27, 28]    # Tokyo
]
```

```python
# Using the Objec Oriented Approach :

fig,axes = plt.subplots(2,2)

axes  = axes.flatten()

for i,ax in enumerate(axes):
    ax.plot(days,temperatures[i],marker="o")
    ax.set_title(cities[i])
    ax.grid()

fig.suptitle("Weekly Temperature by City")
fig.supylabel("Temperature (°C)")
fig.supxlabel("Day")
fig.tight_layout()
```

```python
# My Approach :

cities = [["New York", "London"], ["Delhi", "Tokyo"]]

temperatures = [
    [[22, 23, 21, 24, 25],   # New York
    [18, 19, 17, 20, 21]],   # London
    [[30, 32, 31, 33, 34],   # Delhi
    [25, 26, 24, 27, 28]]    # Tokyo
]

fig,ax = plt.subplots(2,2)

for i in range(0,2):
    for j in range(0,2):
        ax[i][j].plot(days,temperatures[i][j],marker="o")
        ax[i][j].grid(True)
        ax[i][j].set_title(cities[i][j])

fig.suptitle("Weekly Temperature by City")
fig.supylabel("Temperature (°C)")
fig.supxlabel("Day")
fig.tight_layout()
```
