```python
import matplotlib.pyplot as plt
```

```python
# Data :

# Pie Charts
expenses = ["Salaries", "Rent", "Marketing", "R&D", "Miscellaneous"]
amounts = [500, 150, 120, 100, 50]
```

```python
plt.pie(amounts,labels=expenses)
```

```python
plt.pie(amounts,labels=expenses)

# plt.style.available

plt.style.use('dark_background')
```

```python
# Customizations :

# Custom Colours :
colours = ["red","pink","purple","green","orange"]


# wedge props : (edge,linstyle,linewidth,aplha,fill)

# Explode = Take out the slice 
exp = [0.1,0,0,0,0]

# startangle : rotate the pie chart

# Displaying the percentage : ("fill" : false (remove colour))
plt.pie(amounts,labels=expenses,autopct="%1.1f%%",colors=colours,
       wedgeprops={
           "edgecolor": "black",
           "linewidth" : "5",
           "linestyle" : "--",
           "alpha" : 0.9,
       },
        explode=exp,
        startangle=90,
        shadow = True
       )

plt.style.use("classic")
```
