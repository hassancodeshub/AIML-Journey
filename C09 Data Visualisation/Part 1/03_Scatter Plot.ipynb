```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Scatter Plot : Data

people = ["Person A", "Person B", "Person C", "Person D", "Person E", 
          "Person F", "Person G", "Person H", "Person I", "Person J"]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
bp = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]
```

```python
plt.scatter(age,bp)
plt.title("Age Vs Blood Pressure")
plt.xlabel("Age")
plt.ylabel("bp")
plt.grid()
plt.show()

print("Conclusion :")
print("Person H is outlier")
```

```python
# Customization :




# Colour Size Marker Transparency :

colors = ["green" if x<135 else "red" for x in bp]
plt.scatter(age,bp,color=colors,s=bp,marker=">",alpha=0.8)


plt.title("Age Vs Blood Pressure")
plt.xlabel("Age")
plt.ylabel("bp")
plt.grid()
plt.show()

print("Conclusion :")
print("Person H is outlier")
```

```python
# Customization :




# Colour Map :


plt.scatter(age,bp,cmap="plasma",c=bp,s=200)


plt.title("Age Vs Blood Pressure")
plt.xlabel("Age")
plt.ylabel("bp")
plt.grid()
plt.colorbar(label="bp")
plt.show()

print("Conclusion :")
print("Person H is outlier")
```

```python
# Annotation :

plt.scatter(age,bp,cmap="plasma",c=bp,s=50)

for i in range(len(age)):
    plt.annotate(people[i],xy=(age[i],bp[i]),xytext=(age[i]-2,bp[i]+2))

# xticks & yticks ;

plt.xticks(np.arange(15,75,5))
plt.yticks(np.arange(110,160,5))

plt.title("Age Vs Blood Pressure")
plt.xlabel("Age")
plt.ylabel("bp")
plt.grid()
plt.colorbar(label="bp")
plt.xlim(15,70)
plt.ylim(105,160)
plt.show()

print("Conclusion :")
print("Person H is outlier")
```
