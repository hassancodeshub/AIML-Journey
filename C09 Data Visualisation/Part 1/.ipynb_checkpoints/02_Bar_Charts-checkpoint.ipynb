```python
import matplotlib.pyplot as plt
```

```python
# Data : 

# Oscar VS Non Oscar Movies Revenue :

years = [2008, 2009, 2010, 2011, 2012]

# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]
oscar_revenue = [1005, 170, 427, 133, 232]

# Non-Oscar Winner Movies
non_oscar_movies = [
    "Slumdog Millionaire",
    "Avatar",
    "Inception",
    "Hugo",
    "Lincoln"
]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M
```

```python
# Vertical Bar Chart :

plt.bar(years,oscar_revenue,color="purple")
plt.title("Oscar Movies Yearly Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue")
```

```python
# Vertical Bar Chart : With plt.text()

# xlim() & ylim() to change the graph size

plt.bar(years,oscar_revenue,color="green")
plt.title("Oscar Movies Yearly Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue")

for i in range(len(years)):
    plt.text(years[i],oscar_revenue[i]+30,str(oscar_revenue[i]),ha="center")

plt.ylim(0,max(oscar_revenue)+200)
```

```python
# Compare 2 Sets of data : (Overlaped)

plt.bar(years,oscar_revenue,color="green")
plt.bar(years,non_oscar_revenue,color="blue")
plt.title("Oscar Movies Yearly Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue")

plt.ylim(0,max(oscar_revenue)+200)
```

```python
#  Solution : Change the origin of both the charts & change the width as well (Default : 0.8)

w = 0.4 # Width

oscar_revenue = [1005, 170, 427, 133, 232]
non_oscar_revenue = [378, 2788, 829, 185, 275]
years = [2008, 2009, 2010, 2011, 2012]

# Shifting center to the left :
plt.bar([y-w/2 for y in years],oscar_revenue,color="blue",width=w,label="Oscar Movies") 
# Shifting center to the right :
plt.bar([y+w/2 for y in years],non_oscar_revenue,color="orange",width=w,label="Non-Oscar Movies")

"""
plt.bar(x-width/2, oscar_revenue, width, label="Oscar")
plt.bar(x+width/2, non_oscar_revenue, width, label="Non-Oscar")
"""

plt.title("Oscar Movies Yearly Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue")
plt.legend()
```

```python
# Practice 3 Data Sets :

oscar_revenue = [1005, 170, 427, 133, 232]
non_oscar_revenue = [378, 2788, 829, 185, 275]
double_oscar_revenue = [500,364,521,1800,125]
years = [2008, 2009, 2010, 2011, 2012]

w = 0.3 # Width

oscar_revenue = [1005, 170, 427, 133, 232]
non_oscar_revenue = [378, 2788, 829, 185, 275]
years = [2008, 2009, 2010, 2011, 2012]


plt.bar([y-w for y in years],oscar_revenue,color="blue",width=w,label="Oscar Movies") 
plt.bar(years,non_oscar_revenue,color="orange",width=w,label="Non-Oscar Movies")
plt.bar([y+w for y in years],double_oscar_revenue,color="green",width=w,label="Non-Oscar Movies")

plt.title("Oscar Movies Yearly Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue")

plt.legend()
```

```python
# X ticks:

oscar_revenue = [1005, 170, 427, 133, 232]

plt.bar([0,1,2,3,4],oscar_revenue,color="green")

plt.show()

plt.bar([0,1,2,3,4],oscar_revenue,color="green")
plt.xticks([0,1,2,3,4],years)

plt.show()
```

```python
# Horizontal Bar Charts :

"""
plt.barh(categories,data)
"""

plt.barh(years,oscar_revenue)
plt.title("Oscar Movies Yearly Revenue")
plt.xlabel("Revenue")
plt.ylabel("Years")

for i in range(len(years)):
    plt.text(oscar_revenue[i]+70,years[i],str(oscar_revenue[i]),rotation=-90,va="center")

plt.xlim(0,1200)
```
