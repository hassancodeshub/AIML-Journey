```python
# Movies Data

# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]
oscar_years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

plt.plot(oscar_years,oscar_revenue)
plt.title("Oscar Movies Revenue Data Yearly")
plt.xlabel("Years")
plt.ylabel("Revenue in Million")
```

```python
# Oscar VS Non Oscar Movies Revenue :

years = [2008, 2009, 2010, 2011, 2012]

# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]
oscar_revenue = [1005, 170, 427, 133, 232]

# Non-Oscar Winner Movies
non_oscar_movies = [
    "Slumdog Millionaire",
    "Avatar",
    "Inception",
    "Hugo",
    "Lincoln"
]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

# Format String : '[marker][format][colout]'
plt.plot(years,oscar_revenue,'o--r',label="Oscar Movies")
plt.plot(years,non_oscar_revenue,'x-g',label="Non Oscar Movies")
plt.title("Oscar VS Non Oscar Movies Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue in Million")
plt.legend(loc="upper right")  # Default : Upper Right
plt.show()
```

```python
# tight-layout :

plt.plot(oscar_movies,oscar_revenue,label="Oscar Movies")
plt.title("tight-layout")
plt.xlabel("Movies")
plt.ylabel("Revenue in Million")
plt.legend(loc="upper right")  # Default : Upper Right

plt.tight_layout()
plt.savefig("data.png")

plt.show()
```
